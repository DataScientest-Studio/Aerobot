## Branching

### Create a new branch

```python
# To create a new branch and switch to it at the same time, you can run the git checkout command with the -b switch
! git checkout -b new_branch_1

# This is shorthand for:
# ! git branch new_branch_name
# ! git checkout new_branch_name
```

Now you can work on your branch, i.e. modify / create files etc.

For more details on Git Branching - Basic Branching and Merging:
see https://git-scm.com/book/en/v2/Git-Branching-Basic-Branching-and-Merging


### Commit & push the branch

Once you finished your work, commit and push, as you would do with single files, in order to 'save' the branch (and work on it later)

```python
# Commit the branch 
! git commit -m "Branch 1 modifications"
```

```python
!git push --set-upstream origin new_branch_1
```

### Merge the branch with the main branch

Once you have fullfilled the scope of your branch, you can merge it with the main branch (and then delete the branch).

```python
# Merge new_branch_1 to the main branch
! git checkout main # switch to the main branch
! git merge new_branch_1
```

```python
# Swith to another branch
!git checkout new_branch_1
```
